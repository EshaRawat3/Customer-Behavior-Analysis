```python
# handling missing values
# here we are taking the median of reviews based on category like if it is clothing then the median of clothing category will be added in place of null values

df["Review Rating"] =df.groupby('Category')["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# to make it consistent data, easier to read, and snake (all lowercase) will help you to work smoothly with databases (SQL)

df.columns=df.columns.str.lower()  # lower
df.columns=df.columns.str.replace(" ","_")   # replacing spaces to _
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})   #rename the column
```

```python
df.columns
```

```python
#create a column age_group (young, adult, middle-age, senior)

labels=["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"]=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
#create columns purchase_freduency_days
frequency_mapping={
    "Fortnightly": 14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":120,
    "Bi-Weekly": 14,
    "Annually":365,
    "Every 3 Months":90
}

df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```

```python
df[["discount_applied","promo_code_used"]].head(10)
```

```python
# to check whether both contains same
(df["discount_applied"]==df["promo_code_used"]).all()
```

```python
# as they contain same data we are gonna remove "promo code used"
df=df.drop("promo_code_used",axis=1)
```
